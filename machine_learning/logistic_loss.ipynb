```python
# Logistic regression loss function
# This script defines the logistic loss function and plots it for both classes (y_true = 1 and y_true = 0).
import numpy as np
import matplotlib.pyplot as plt

def logistic_loss(y_true, y_pred):
    """
    Compute the logistic loss function for binary classification.
    This function calculates the logistic loss for binary classification problems.
    
    Parameters:
    y_true : array-like, shape (n_samples,)
        True binary labels (0 or 1).
    y_pred : array-like, shape (n_samples,)
        Predicted probabilities for the positive class (between 0 and 1).
        
    Returns:
    loss : float
        The computed logistic loss.
    """
    loss = -y_true * np.log(y_pred) - (1 - y_true) * np.log(1 - y_pred)
    return loss

def plot_logistic_loss(y_pred, loss_y1, loss_y0):
    """
    Plot the logistic loss function for both classes (y_true = 1 and y_true = 0).
    """
    plt.figure(figsize=(10, 6))
    plt.plot(y_pred, loss_y1, label='y_true = 1', color='blue')
    plt.plot(y_pred, loss_y0, label='y_true = 0', color='red')
    plt.title('Logistic Loss Function')
    plt.xlabel('Predicted Probability (y_pred)')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.show()


if __name__ == "__main__": 
    # Generate predicted probabilities
    y_pred = np.linspace(0.01, 0.99, 100)  # Avoiding 0 and 1 to prevent log(0)
    # Calculate losses for y_true = 1 and y_true = 0
    loss_y1 = logistic_loss(np.ones_like(y_pred), y_pred)
    loss_y0 = logistic_loss(np.zeros_like(y_pred), y_pred)
    # Plot the logistic loss
    plot_logistic_loss(y_pred, loss_y1, loss_y0)
```
