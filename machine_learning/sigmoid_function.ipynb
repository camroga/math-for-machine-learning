```python
# sigmoide function
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
    Sigmoid function
    :param x: input value or array
    :return: sigmoid of x
    """
    return 1 / (1 + np.exp(-x))

# plot the sigmoid function
def plot_sigmoid(X, y):
    """
    Plot the sigmoid function
    :param X: input values
    :param y: sigmoid values
    """
    plt.figure(figsize=(8, 4))
    plt.plot(X, y, label='sigmoid', color='blue')
    plt.title('Sigmoid Function')
    plt.xlabel('Input value')
    plt.ylabel('Sigmoid output')
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    plt.grid()
    plt.legend()
    plt.show()

    
# main function to test the sigmoid function
if __name__ == "__main__":
    # Test the sigmoid function with a range of values
    test_values = np.array([-10, -5, 0, 5, 10])
    results = sigmoid(test_values)
    plot_sigmoid(test_values, results)
    
    # Print the results
    for value, result in zip(test_values, results):
        print(f"sigmoid({value}) = {result}")
```
