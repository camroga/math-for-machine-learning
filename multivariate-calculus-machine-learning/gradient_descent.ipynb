```python
# Gradient descent for f=y*x^2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Function to compute the gradient of f
def gradient(x, y):
    """Compute the gradient of f = y * x^2."""
    dfdx = 2 * y * x
    dfdy = x**2
    return np.array([dfdx, dfdy])

# Function to compute the function value
def f(x, y):
    """Compute the function value f = y * x^2."""
    return y * x**2

# Gradient descent function
# starting_point (sp) = (1, 1)
# learning_rate (lr) = 0.1
# num_iterations (ni) = 100
def gradient_descent(sp, lr, ni):
    """Perform gradient descent."""
    x, y = sp
    trajectory = [(x, y)]
    for _ in range(ni):
        grad = gradient(x, y)
        x -= lr * grad[0]
        y -= lr * grad[1]
        trajectory.append((x, y))
    return np.array(trajectory)

# main function
def main():
    # Create grid for surface and gradient
    x_vals = np.linspace(-2, 2, 20)
    y_vals = np.linspace(-2, 2, 20)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = f(X, Y)

    # Run gradient descent
    start = (-1.5, 1.5)
    lr = 0.1
    steps = 10
    trajectory = gradient_descent(start, lr, steps)
    print("Gradient Descent Trajectory:")
    print(trajectory)
    x_traj, y_traj = trajectory[:, 0], trajectory[:, 1]
    z_traj = f(x_traj, y_traj)

    # Plotting
    fig = plt.figure(figsize=(12, 8))

    ax = fig.add_subplot(111, projection='3d')
    ax.view_init(elev=0, azim=-70)

    # Plot the surface
    ax.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.8)
    
    # Plot the gradient descent path
    ax.plot(x_traj, y_traj, z_traj, color='black', linewidth=1, marker='.', label='Gradient Descent Path')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x, y)')
    ax.set_title('Gradient Descent on 3D Surface')
    ax.legend()
    plt.tight_layout()
    plt.show()
    
if __name__ == "__main__":
    main()



```
