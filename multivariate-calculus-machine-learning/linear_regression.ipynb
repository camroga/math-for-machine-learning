```python
# Linear regression doing a manual implementation

import numpy as np
import matplotlib.pyplot as plt

def linfit(xdat,ydat):
  # Here xbar and ybar are calculated
  xbar = np.sum(xdat)/len(xdat)
  ybar = np.sum(ydat)/len(ydat)
  print(f"xbar: {xbar}, ybar: {ybar}")

  # Insert calculation of m and c here. If nothing is here the data will be plotted with no linear fit
  # Calculate m (slope)
  numerator = np.sum((xdat - xbar) * (ydat - ybar))
  denominator = np.sum((xdat - xbar) ** 2)
  m = numerator / denominator
  c = ybar -m*xbar 

  # Return your values as [m, c]
  return [m, c]

def generate_data():
    # Generate some data
    np.random.seed(0)
    X = 5 * np.random.rand(50, 1)
    y = 4 + 3 * X**2 + np.random.randn(50, 1)
    return X, y

def main():
    # Generate data
    X, y = generate_data()
    # Call the function
    m, c = linfit(X, y)
    # Print the values of m and c
    print(f"m: {m}, c: {c}")
    # Plot the data and the linear fit
    plt.scatter(X, y)
    plt.plot(X, m * X + c, color='red', label='Linear Fit')
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title("Linear Fit")
    plt.legend()
    plt.show()

# Call the main function
if __name__ == "__main__":
    main()
```
