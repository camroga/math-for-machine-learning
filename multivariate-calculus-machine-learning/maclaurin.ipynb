```python
#Using Maclaurin's formula to calculate y = x^3 + x^2 + x + 1
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp


function = sp.sin(sp.symbols('x'))
x = sp.symbols('x') 
x0 = 0
n = 3
# Maclaurin's formula
# f(x) = sen(x)
f = function.subs(x, x0)
# f'(x) = cos(x)
f1 = function.diff(x).subs(x, x0)
# f''(x) = -sen(x)
f2 = function.diff(x, 2).subs(x, x0)
# f'''(x) = -cos(x)
f3 = function.diff(x, 3).subs(x, x0)
# f''''(x) = sen(x)
f4 = function.diff(x, 4).subs(x, x0)
# f'''''(x) = cos(x)
f5 = function.diff(x, 5).subs(x, x0)
maclaurin = f + f1*x + f2*x**2/2 + f3*x**3/6 + f4*x**4/24 + f5*x**5/120
# Displaying the results
print(f)
print(f1)
print(f2)
print(f3)
print(f4)
print(maclaurin)
# Plotting the function and the Maclaurin series
x_vals = np.linspace(-3, 3, 100)
y_vals = [function.subs(x, val) for val in x_vals] 
maclaurin_vals = [maclaurin.subs(x, val) for val in x_vals]
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, maclaurin_vals, label='Maclaurin series')
plt.title('Maclaurin Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
