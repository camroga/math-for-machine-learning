```python
#Using Maclaurin's formula to calculate y = x^3 + x^2 + x + 1
# f(x) = sen(x)
# f'(x) = cos(x)
# f''(x) = -sen(x)
# f'''(x) = -cos(x)

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp


function = sp.sin(sp.symbols('x'))
x = sp.symbols('x') 
x0 = 0
n = 3
# Maclaurin's formula  
maclaurin = function.subs(x, x0) + function.diff(x).subs(x, x0)*x + function.diff(x, 2).subs(x, x0)*(x)**2/2 + function.diff(x, 3).subs(x, x0)*(x)**3/6
print(maclaurin)
# Plotting the function and the Maclaurin series
x_vals = np.linspace(-2, 2, 100)
y_vals = [function.subs(x, val) for val in x_vals] 
maclaurin_vals = [maclaurin.subs(x, val) for val in x_vals]
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, maclaurin_vals, label='Maclaurin series')
plt.title('Maclaurin Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
