```python
#Using Taylor series
#to approximate the value of y= sin(xy/5)
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from mpl_toolkits.mplot3d import Axes3D

# Define symbols
x, y = sp.symbols('x y')

# Define the function
function = sp.sin(x * y / 5)
x0 = 1
y0 = 1

# First order Taylor expansion around (x0, y0)
f = function.subs({x: x0, y: y0})
f_x = function.diff(x).subs({x: x0, y: y0})
f_y = function.diff(y).subs({x: x0, y: y0})

# Taylor series up to first order
taylor = f + f_x * (x - x0) + f_y * (y - y0)

# Convert to numerical functions
taylor_func = sp.lambdify((x, y), taylor, 'numpy')
original_func = sp.lambdify((x, y), function, 'numpy')

# Create grid
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluate functions
Z = original_func(X, Y)
Z_taylor = taylor_func(X, Y)

# Plotting
fig = plt.figure(figsize=(14, 6))

# Original function - 3D plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Original Function')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# Taylor approximation - 3D plot
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_taylor, cmap='plasma')
ax2.set_title('Taylor Approximation (1st Order)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('T(x, y)')

plt.tight_layout()
plt.show()





```
