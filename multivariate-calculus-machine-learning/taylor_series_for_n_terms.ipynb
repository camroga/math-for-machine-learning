```python
# Taylor series for f(x) = x sin(x)
# Taylor series expansion around x = 0
# f(x) = x sin(x)
# f(0) = 0
# f'(0) = 1
# f''(0) = 0
# f'''(0) = -1
# f''''(0) = 0
# f'''''(0) = 1
# f''''''(0) = 0
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


# Step 1: Define symbol and function
x = sp.symbols('x')
function = x * sp.sin(x)

# Step 2: Define the expansion point (Maclaurin = x0 = 0)
x0 = 0
n = 6  # Degree of the Taylor (Maclaurin) series

# Step 3: Initialize the series with f(x0)
# This is the 0th-order term
f = function.subs(x, x0)
taylor_series = f


# Step 4: Build higher-order terms (from 1 to n)
for i in range(1, n + 1):
    # Compute i-th derivative
    derivative = function.diff(x, i)

    # Evaluate at x0
    derivative_at_x0 = derivative.subs(x, x0)
    
    # Compute i-th term of the series
    term = (derivative_at_x0 / sp.factorial(i)) * (x - x0)**i
    
    # Add term to the series
    taylor_series += term
    
    # Print each term clearly
    print(f"f^{i}({x0}) = {derivative_at_x0}, term = {term}")

# Plot the function and its Taylor series approximation
x_vals = np.linspace(-4, 4, 100)
f_vals = [function.subs(x, val) for val in x_vals]
taylor_vals = [taylor_series.subs(x, val) for val in x_vals]
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = x sin(x)', color='blue')
plt.plot(x_vals, taylor_vals, label=f'Taylor Series (degree {n})', color='red', linestyle='--')
plt.title('Taylor Series Approximation of f(x) = x sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()



```
