```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# calculate the mean, variance, and covariance of a dataset
def calculate_statistics(data):
    mean = np.mean(data, axis=0)
    variance = np.var(data, axis=0)
    covariance = np.cov(data, rowvar=False)
    return mean, variance, covariance

# mock dataset
def generate_data():
    data = np.array([[1,1],
                [-3,-3],
                [-1,1],
                [2,2],
                [0,1],
                [0,2],
                [-2,0],
                [1,3],
                [4,4]])
    return data

# plot the dataset
def plot_data(data, mean, variance, covariance):
    plt.figure(figsize=(10, 6))
    plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
    plt.title('Random Data Scatter Plot')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # plot mean
    plt.scatter(mean[0], mean[1], color='red', label='Mean', s=100)
    
    # plot variance in a line
    plt.plot([mean[0] - np.sqrt(variance[0]), mean[0] + np.sqrt(variance[0])],
             [mean[1], mean[1]], color='green', label='Variance', linewidth=2)
    plt.plot([mean[0], mean[0]],
             [mean[1] - np.sqrt(variance[1]), mean[1] + np.sqrt(variance[1])], color='green', linewidth=2)
    

    # plot covariance showing the direction of the data spread
    plt.quiver(mean[0], mean[1], covariance[0, 0], covariance[1, 0], angles='xy', scale_units='xy', scale=1, color='blue', label='Covariance')
    plt.legend()
    plt.grid()
    plt.show()
    
if __name__ == "__main__":
    # Generate random data
    data = generate_data()

    # Calculate statistics
    mean, variance, covariance = calculate_statistics(data)

    # Plot the data
    plot_data(data, mean, variance, covariance)
```
