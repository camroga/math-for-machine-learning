```python
# projection 1D matrix in the subspace with base b
import numpy as np
import matplotlib.pyplot as plt

def project_1d_matrix(x, b):
    """
    Projects a 1D matrix A onto the subspace spanned by the vector b.
    
    Parameters:
    A (numpy.ndarray): The input 1D matrix (vector).
    b (numpy.ndarray): The base vector defining the subspace.
    
    Returns:
    numpy.ndarray: The projection of A onto the subspace spanned by b.
    """
    b_norm = np.linalg.norm(b)
    if b_norm == 0:
        raise ValueError("Base vector b cannot be zero.")
    
    PM = np.outer(b, b) / np.power(np.linalg.norm(b), 2)
    p = PM @ x
    return p

def plot_projection(x, b):
    """
    Plots the projection of a 1D matrix A onto the subspace spanned by b.
    
    Parameters:
    A (numpy.ndarray): The input 1D matrix (vector).
    b (numpy.ndarray): The base vector defining the subspace.
    """
    p = project_1d_matrix(x, b)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, label='Original Vector x', marker='o')
    plt.plot(p, label='Projection of x onto b', marker='x')
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.title('Projection of x onto Subspace Spanned by b')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == "__main__":
    x = np.array([1, 2, 3])
    b = np.array([0, 1, 0])
    projection = project_1d_matrix(x, b)
    plot_projection(x, b)
    print("Projection of x onto the subspace spanned by b:", projection)
    
```
